# Regressione pesata

La regressione pesata è una tecnica che viene utilizzata quando i dati non sono omoschedastici, ovvero quando la varianza degli errori non è costante per tutte le osservazioni. I modo per individuare sintomi di eteroschedasticità sono:
- Individuare se i dati derivano da una distribuzione di Poisson in cui la varianza è approssimativamente uguale alla media ($\sigma^2_i \approx \mu_i$).
- Individuare se i dati derivano da una distriubuzione Binomiale in cui la varianza è di una proporzione dipende dalla proporzione stessa.
- Verificare se nei grafici dei residui si mostra una forma a "imbuto", essa indica che la dispersione di residui cambia al variare dei valori stimati.

Quando l'assunzione di omoschedasticità cade ogni osservazione $Y_i$ segue una distribuzione Normale con una propria varianza: $Y_i \sim N(\mu_i, \sigma_i^2)$. All'interno della regressione lineare avevamo che per trovare gli stimatori andavamo ad applicare il principio di Massima Verosomiglianza assumendo che la densità di probabilità di ogni previsione fosse:

$$f(y_i;\,\beta_0,\beta_1,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\!\left(-\frac{(y_i - \beta_0 - \beta_1 x_i)^2}{2\sigma^2}\right)$$

quindi mantenevamo una varianza costante tra tutti i dati, ora riscrivendo la formula come:

$$f(y_i;\,\beta_0,\beta_1,\sigma^2_i) = \frac{1}{\sqrt{2\pi\sigma^2_i}}\exp\!\left(-\frac{(y_i - \beta_0 - \beta_1 x_i)^2}{2\sigma^2_i}\right)$$

Andiamo a esplicitare la varianza dell'errore sull'i-esima predizione effettuata. Abbiamo che la formula di log-verosomiglianza, già risolta e già adattata per casi multipli, si modificherà in:
$$
l(β)=C - \sum_{i=1}^n\frac {(Y_i-\sum_{j=0}^p \beta_j x_{ij})^2} {2\sigma_i^2}
$$

Dove C sono i termini costanti che possiamo ignorare, quindi per massimizare $l(\beta)$ ci basterà risolvere il seguente problema di minimizzazione:
$$
\min_{\beta} (\sum_{i=1}^n \frac 1 {\sigma_i^2} (Y_i-\sum_{j=0}^p \beta_j x_{ij})^2)
$$

questa formula prenderà il nome di **minimi quadrati pesati (WLS)** e i pesi $w_i$ emergeranno dalla formula:
$$
w_i = \frac 1 {\sigma_i^2}
$$

Così facendo ciò che accadrà è che le osservazioni più precise (quindi con varianza più piccola) avranno un peso maggiore, mentre quelle più imprecise avranno un peso minore.

## Stima dei pesi

Notare che la varianza legata all'i-esima predizione è praticamente impossibile da trovare, perciò tipicamente la tecnica che si utilizza è quella di:
1. Calcolare i residui derivati dalla OLS sui dati.
2. Fare il plot del grafico dei residui e verificare in quale situazione ci troviamo:
   1. $\sigma_i \propto \hat Y_i$: ovvero la varianza dei residui è proporzionale all'aumento del valore di $Y_i$ (tipicamente grafico ad imbuto), allora possiamo calcolare i pesi come:
    $$w_i = \frac 1 {\hat Y_i^2}$$
   2. $\sigma_i \propto \sqrt{\hat Y_i}$: ovvero la varianza dei residui è proporzionale all'aumento del valore della radice quadrata di $\hat Y_i$, allora possiamo calcolare i pesi come:
    $$w_i = \frac 1 {Y_i}$$
    Notare che in entrambi i casi $\hat Y_i$ indicano valori predetti.
