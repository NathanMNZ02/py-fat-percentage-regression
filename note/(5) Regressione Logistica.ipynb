# Regressione Logistica

La regressione logistica è un modello statistico utilizzato quando la variabile di risposta Y è categoriale. In tal caso a differenza della regressione lineare non modella il valore della risposta, ma piuttosto la probabilità che la variabile appartenga a una determinata categoria.

## Regressione Logistica Binomiale

Utilizzata quando la variabile di risposta è **dicotomica**, ovvero può assumere solo due valori, es:
- 0/1.
- successo/fallimento.
- malato/sano.
- ...

L'obiettivo sarà quindi è modellare la probabilità sia 1 utilizzando una **funzione logistica (sigmoide)**, ovvero una funzione che mappa i predittori in un intervallo [0, 1]:
$$
p(X) = \sigma(z) = \frac 1 {1+e^{-z}}
$$

dove $z = \beta_0+\beta_1 x_1 + \dots + \beta_p x_p$. In tal caso la stima dei parametri sarà leggermente differente, ovvero in un problema di regressione binomiale stiamo assumendo che ogni osservazione $y_i \in \{0,1\}$ segua una Bernoulli con paraemtro $p_i = \sigma(z_i) = P(y_i = 1 | x_i)$. Quindi la funzione di massa di $y_i$ sarà descritta come:
$$
P(y_i = 1 | x_i) = p_i \quad P(y_i = 0 | x_i) = 1 - p_i \quad \Rightarrow P(y_i | x_i) = p_i^{y_i} (1-p_i)^{1-y_i}
$$

Assumendo le osservazioni tra di loro indipendenti avremmo che la funzione di log verosomiglianza (vista nella regressione lineare per la stima dei parametri) diventerà:

$$
l(\beta) = \sum_{i=1}^n [y_i \log p_i + (1-y_i)\log(1-p_i)]
$$

In tal caso avremmo che massimizzare $l(\beta)$ equivale a minimizzare la negative log verosomiglianza che in machine learning essa equivale alla **cross-entropy**.

## Regressione Logistica Multinomiale

Generalizzazione del modello a casi in cui la variabile di risposta ha più di due categorie, in tal caso il modello dovrà predirre un vettore di probabilità $\pi = (\pi_1, \pi_2, \dots, \pi_m)$ dove $\pi_i$ è la probabilità che un'osservazione appartenza alla categoria $\pi_i$, notare che $\sum \pi_i = 1$. Per predirre questo vettore di probabilità si utilizza la **funzione softmax**, ovvero una generalizzazione della funzione sigmoide:
1. Per ogni categoria $i$ si calcola un logit $z_i$ con un set di parametri dedicato: $z_i=w_{i0} + w_{i1}x_1+\dots+w_{ip}x_p$.
2. La probabilità relativa alla categoria $i$ sarà quindi calcolata tramite la softmax che normalizza i logit:
    $\pi_i = \frac {e^{z_i}} {\sum_{j=1}^m e^{z_j}}$

Anche in uqesto caso la stima dei parametri $w$ avverrà andando a minimizzare la cross entropy, la quale per una singola osservazione sarà:
$$
H(y_i, \pi_i) = - \sum_{k=1}^m y_{ik} \log(\pi_{ik})
$$
