# Design of Experiments (DoE)

Il Design of Experiments è una metodologia statistica che serve a pianificare, condurre e analizzare esperimenti in modo efficiente, così da capire quali fattori influenzano davvero un fenomeno e in che modo. Esso viene utilizzato per:

- Identificare variabili che influenzano un risultato.
- Misurare l'effetto di ogni variabile.
- Scoprire interazioni tra variabili.
- Ottimizzare i processi.

## Inferenza statistica

Quando osserviamo dei dati non vediamo mai la realtà completa di una popolazione, ma solo un campione. Questo introduce un punto fondamentale:
> I dati sono sempre affetti da variabilità casuale.

Quindi il problema dell'inferenza è _"Come posso prendere decisioni su una popolazione usando solo un campione rumoroso?"_. Il **test di ipotesi** è la procedura per prendere decisioni in condizioni di incertezza, esso funziona come un processo in tribunale:
- $H_0$: l'ipotesi nulla e che vogliamo smentire (l'imputato).
- $H_1$: l'ipotesi alternativa che si contrappone all'ipotesi nulla.
- I dati campionari: sono le prove presentate.
- Lo statistico: è il giudice che valuta se le prove sono abbastanza schiaccianti per rifiutare l'ipotesi nulla.

Lo statistico dovrà calcolare:
- Statistica test: quanto i dati si discostano dall'ipotesi nulla $H_0$, più si discostano più $H_0$ diventa sospetta.
- P-value: la probabilità di osservare un risultato così estremo o più estremo nell'ipotesi che $H_0$ sia vera. Ciò significa determinare quanto sono strani i dati nell'ipotesi che $H_0$ sia vera.

La decisione dello statistico per determinare se $H_0$ sia vera o meno si basa su una regola decisionale dettata dal valore del p-value, si introduce quindi una **soglia di significatività** $\alpha$ (solitamente 0.05):
- Se $\text{p-value} < \alpha$ si rifiuta $H_0$.
- Se $\text{p-value} \ge \alpha$ non si rifiuta $H_0$, non abbiamo prove sufficienti per smentirla.

### Errori e potenza di un Test

Nel prendere una decisione basata su un test di ipotesi ci sono quattro possibili esiti:
- Non rifiuto $H_0$:
  - Se $H_0$ è vera allora abbiamo una decisione corretta.
  - Se $H_0$ è falsa allora abbiamo un **errore di tipo 1 (falso positivo)** $\alpha$.
- Rifiuto $H_0$:
  - Se $H_0$ è falsa allora abbiamo una decisione corretta.
  - Se $H_0$ è vera allora abbiamo un **errore di tipo 2 (falso negativo)** $\beta$.

La probabilità di rifiutare correttamente un'$H_0$ falsa prende il nome di **potenza** e viene calcolata come:
$$
\text{Potenza}= 1- \beta
$$
