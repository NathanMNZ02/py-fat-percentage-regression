# ANOVA

L'analisi della varianza (ANOVA) è una tecnica statistica che può essere vista come una variante della regressione lineare utilizzata quando le variabili in ingresso sono **categoriali**. Il modello ANOVA analizza quindi la relazione tra una o più variabili categoriali indipendenti e una variabile dipendente numerica assumendo che l'errore sia additivo, Gaussiano e omoschedastico.

## ANOVA a Una Via

Permette di verificare se una singola variabile categoriale influenza una variabile numerica, confrontando le medie di più gruppi. Suppponiamo ad esempio:
- Variabile categoriale (fattore) = metodo di studio.
  - A
  - B
  - C
- Variabile numerica = voto esame.

Con ANOVA possiamo rispondere alla domanda:

"I voti medi dei tre metodi sono davvero diversi oppure la differenza è casuale?"

L'idea centrale è che ogni categoria della variabile d'ingresso possa avere una propria media per la variabile di risposta, supponiamo $Y_{ij}$ la $j$-esima osservazione nel $i$-esimo gruppo. Allora il modello statistico è:
$$
Y_{ij} \sim N(\mu_i, \sigma^2)
$$

Quindi i parametri incogniti del modello saranno le medie dei gruppi $\mu_1, \mu_2, \dots, \mu_m$ e la varianza comune $\sigma^2$. La stima dei parametri risulta banale di fatti basterà fare:
- Stima delle Medie ($\mu_i$): 
    $$\hat \mu_i = \overline Y_i = \frac 1 {n_i} \sum_{j=1}^{n_i} Y_{ij}$$

- Stima della Varianza Comune ($\sigma^2$): 
    1. Per prima cosa si calcola la varianza campionaria di ciascun gruppo $i$:
        $$S_i^2 = \frac 1 {n_i - 1} \sum_{j=1}^{n_i} (Y_{ij} - \overline Y_i)^2$$

    2. Per ottenere una singola stima robusta si esegue una media pesata sulle $m$ varianze, lo stimatore risultante è lo **stimatore pooled della varianza** o **Within** o **Mean Square Within** ($S^2_W$):
        $$S^2_W = \frac {\sum_{i=1}^m (n_i - 1)S_i^2} {N-m}$$

        Con $N$ il numero totale di osservazioni.

Ora per testare le medie dei vari gruppi costruiamo le seguenti ipotesi:
- **Ipotesi Nulla ($H_0$)**: tutte le medie dei gruppi sono uguali:
    $$H_0: \mu_1 = \mu_2 = \dots = \mu_m$$
    Sotto $H_0$ la variabile categoriale non ha alcun effetto sulla risposta $Y$.

- **Ipotesi Alternativa ($H_1$)**: non tutte le medie sono uguali (almeno una è diversa). Sotto $H_1$ la variabile categoriale ha effetto su $Y$.

Per confrontare quindi la variabilità tra i gruppi utilizziamo la seguente statistica test:
$$
F = \frac {S^2_B} {S^2_W}
$$

Dove $S^2_B$ misura quanto le medie dei gruppi si discostano dalla media generale $\overline Y$ calcolata su tutte le $N$ osservazioni:
$$
S^2_B = \frac {\sum_{i=1}^m n_i(\overline Y_i - \overline Y)^2} {m-1}
$$

L'assunzione che quindi viene fatta è che sotto $H_0$ sia $S^2_B$ che $S^2_W$ stimano la stessa quantità, cioè $\sigma^2$, quindi il rapporto $F$ tende a stare vicino a 1, sotto $H_1$ il numeratore stima $\sigma^2$ più un termine positivo dovuto alla differenza tra le medie, quindi $F$ tende a essere grande. Una volta calcolata la statistica si può calcolare il p-value, ovvero la probabilità sotto $H_0$ di osservare un valore della statistica $F$ grande:

$$
\text{p-value} = 1 - F_{\text{CDF}}(F; m-1, N-m)
$$

Dove $F_{\text{CDF}}$ è la CDF della F di Fisher.

### Equivalenza con la regressione lineare

Notare che ANOVA è equivalente a una regressione lineare con solo variabili dummy in ingresso, di fatti stimando medie e varianza comune stiamo stimando:

$$
Y = \beta_0 + \beta_1 x_1 + \beta x_2 + \dots + \epsilon
$$

## ANOVA a due vie

Date in questo caso DUE variabili categoriali permette di verificare: 
- L'effetto del primo fatto sulla variabile numerica (effetto principale).
- L'effetto del secondo fatto sulla variabile numerica (effetto principale).
- La loro interazione, cioè se l'effetto di un fattore cambia a seconda dei livelli dell'altro.

In questo caso, data $Y_{ijk}$ la $k$-esima osservazione (o replica) per il livello $i$ del primo fattore (righe) e il livello $j$ del secondo fattore (colonne), ANOVA a due vie prevede di scomporre la media della cella $(i,j)$ in modo da separare i diversi effetti. A seconda della presenza o meno di repliche, ovvero di situazioni in cui sono presenti entrambe le categorie $i$ e $j$ distinguiamo due modelli:

- **Modello con interazione** (usato con repliche, $l \ge 2$): scompone la media della cella in quattro parti:
    $$
    \mu_{ij} = \mu + \alpha_i + \beta_j + \gamma_{ij}
    $$
    dove $\mu$ è la media globale, $\alpha_i$ è l'effetto principale del fattore di riga, $\beta_j$ è l'effetto principale del fattore di colonna e $\gamma_{ij}$ è il termine di interazione.

- **Modello additivo** (usato senza repliche, $l = 1$): assume che non ci sia interazione:
    $$
    \mu_{ij} = \mu + \alpha_i + \beta_j
    $$
    In questo modello l'effetto di un fattore è costante a tutti i livelli dell'altro fattore. Per garantire l'unicità del modello si impongono dei vincoli di somma zero sugli effetti ($\sum_i \alpha_i = 0$, $\sum_j \beta_j = 0$).

In entrambi i modelli avremo che gli stimatori di $\alpha_i$, $\beta_i$ e $\gamma_{ij}$ saranno lo scostamento delle medie parziali dalla media generale:
$$
\hat \alpha_i = \overline Y_{i*} - \overline Y_{**} \quad \hat \beta_j = \overline Y_{*j} - \overline Y_{**} \quad \hat \gamma_{ij} =  \overline Y_{ij*} - \overline Y_{i*} - \overline Y_{*j*} + \overline Y_{***}
$$

Ora tramite ANOVA a due vie possiamo testare i vari modelli ponendo le seguenti ipotesi:
- Test sul fattore A (effetto di riga):
  - **Ipotesi Nulla ($H_0^A$)**: 
        $$H_0^A: \alpha_1 = \alpha_2 = \dots = \alpha_a = 0$$
  - **Ipotesi Alternativa ($H_1^A$)**: almeno un $\alpha_i \ne 0$.

- Test sul fattore B (effetto di colonna):
  - **Ipotesi Nulla ($H_0^B$)**: 
        $$H_0^B: \beta_1 = \beta_2 = \dots = \beta_b = 0$$
  - **Ipotesi Alternativa ($H_1^B$)**: almeno un $\beta_i \ne 0$.

- Test sull'interazione (solo se $l>1$):
  - **Ipotesi Nulla ($H_0^{AB}$)**: 
        $$H_0^B: \gamma_1 = \gamma_2 = \dots = \gamma_{ab} = 0$$
  - **Ipotesi Alternativa ($H_1^{AB}$)**: almeno un $\gamma_{ij} \ne 0$.
